```python
import pandas as pd
```

```python
df = pd.read_parquet("yellow_tripdata_2025-01.parquet")
df.head()
print(f"This file contains {len(df)} rows")
```

```python
# Random sample of 10 rows
df.sample(10)
```

```python
# Show columns, types and missing values
df.dtypes
df.isnull().sum()
```

```python
# Mean of the passenger_count to improve analysis
mean_passengers = df['passenger_count'].mean()

# Replace the missing values with the mean
df["passenger_count"] = df["passenger_count"].fillna(mean_passengers)


```

```python
# New dataset with the columns of interest
df = df[["tpep_pickup_datetime", "tpep_dropoff_datetime", "passenger_count", 
         "trip_distance", "PULocationID", "DOLocationID", "payment_type", 
         "fare_amount", "tip_amount", "total_amount"]]
# Random test
df.sample(10)
```

```python
# Handle NaN values and insert random to payment_type
import numpy as np 
payment_mapping = [0, 1, 2, 3, 4, 5, 6]
df["payment_type"] = df["payment_type"].apply(lambda x: np.random.choice(payment_mapping) if pd.isna(x) else x)

# Mapping payment_type to descriptive names
df["payment_type"] = df["payment_type"].map({0: "debit", 1: "credit", 2: "cash", 3: "no charge", 4: "dispute", 5: "unknown", 6: "voided trip"})
```

```python
# Check data types to align with PostgreSQL
df.dtypes
```

```python
# Convert payment_type to a categorical type first, then to string
df["payment_type"] = df["payment_type"].astype("category").astype("string")

# Check data types again
print(df.dtypes)
```

```python
# Rename columns to match the PostgreSQL table
df.rename(columns={"tpep_pickup_datetime": "pickup_datetime", "tpep_dropoff_datetime": "dropoff_datetime"}, inplace=True)
```

```python
#save the new dataset
df.to_parquet('../data/yellow_tripdata_2025-01_clean.parquet')
```
